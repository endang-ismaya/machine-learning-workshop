# Sierpinski Triangle

First Transformation:  
    X<sub>n+1</sub> = 0.5X<sub>n</sub>  
    Y<sub>n+1</sub> = 0.5Y<sub>n</sub>  

Second Transformation:  
    X<sub>n+1</sub> = 0.5X<sub>n</sub> + 0.5  
    Y<sub>n+1</sub> = 0.5Y<sub>n</sub> + 0.5

Third Transformation:  
    X<sub>n+1</sub> = 0.5X<sub>n</sub> + 1  
    Y<sub>n+1</sub> = 0.5Y<sub>n</sub>

```python
import matplotlib.pyplot as plt
from random import choice
# this will show the graph in the jupyter notebook
%matplotlib inline 
```

```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1, y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1, y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y
    return x1, y1

transformations = [trans_1, trans_2, trans_3]
a1 = [0]
b1 = [0]
a, b = 0, 0

for i in range(1_000_000):
    trans = choice(transformations)
    a, b = trans((a, b))
    a1.append(a)
    b1.append(b)
```

```python
plt.rc('figure', figsize=(16, 16))
```

```python
plt.plot(a1, b1, 'o')
plt.savefig('fig-sierpinski-triangle.png')
```
