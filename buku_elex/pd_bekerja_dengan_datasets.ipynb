# Bekerja Dengan Beberapa DataSet

```python
import pandas as pd
```

```python
# dataset
engineering = "http://thinkjubilee.com/data/engineering.csv"
state = "http://thinkjubilee.com/data/state.csv"
ivies = "http://thinkjubilee.com/data/iviesleague.csv"
liberalarts = "http://thinkjubilee.com/data/liberalarts.csv"
wilayah = "http://thinkjubilee.com/data/regions.csv"
```

#### engineering -> daftar universitas yang memang identik dengan dunia engineering
#### State -> daftar universitas negeri yang membuka bidang studi engineering
#### ivies -> daftar universitas yang tergabung dalam kelompok Ivy League yang membuka bidang studi engineering
#### liberalarts -> daftar lembaga pendidikan yang mengembangkan liberal arts untuk bidang studi engineering

```python
pd.read_csv(engineering).head() # membaca hanya 5 row saja
```

```python
pd.read_csv(state).head()
```

```python
pd.read_csv(ivies).head()
```

```python
pd.read_csv(liberalarts).head()
```

```python
# membuat variable dari data set
en = pd.read_csv(engineering)
st = pd.read_csv(state)
ivy = pd.read_csv(ivies)
art = pd.read_csv(liberalarts)
wil = pd.read_csv(wilayah)
school = [en, st, ivy, art]
pd.concat(school)
```

```python
# menggabungkan column axis
pd.concat([en, ivy], axis=1)
```

```python
# menambahkan kolom
en_usa = en.copy()
en_usa["Location"] = "USA"
en_usa
```

```python
# join outer and inner
pd.concat([en_usa, ivy], join="outer")
```

```python
pd.concat([en_usa, ivy], join="inner")
```

## Menggabungkan (Merge) dengan wilayah

```python
wil.head()
```

```python
# merge school
school = [en , st, ivy, art]
sekolah = pd.concat(school)
```

```python
pd.merge(sekolah, wil)
```

```python
# menambahkan kolom yang tidak ada kolom yg sama
income = "http://thinkjubilee.com/data/gaji.csv"
gaji = pd.read_csv(income)
gaji
```

```python
# pd.merge(sekolah, gaji) -> ini akan error karena tidak ada kolom dengan nama yang sama saat di merge
# solusinya adalah dengan menggukanan 'lef_on' dan 'righ_on'
pd.merge(sekolah, gaji, left_on="School Name", right_on="school_name").drop("school_name", axis=1)
```
