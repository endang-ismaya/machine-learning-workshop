# Membuat Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
penjualan = [1000, 3500, 7000, 6000, 6500, 3600, 4500, 5000, 1500, 2000, 1300, 8000]

fig = plt.figure(figsize=(10, 10))
plt.bar(bulan, penjualan)
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Nilai Penjualan")
plt.show()
```

```python
# Subplot
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
penjualan_a = [1000, 3500, 7000, 6000, 6500, 3600, 4500, 5000, 1500, 2000, 1300, 8000]
penjualan_b = [200, 500, 800, 800, 700, 900, 500, 450, 300, 130, 500, 1700]
sumbu_x = np.arange(len(bulan))

fig = plt.figure(figsize=(10, 10))

chart = plt.subplot()
chart.bar(sumbu_x - 0.2, penjualan_a, width=0.4, label="Kacang Goreng")
chart.bar(sumbu_x - 0.3, penjualan_b, width=0.4, label="Kacang Rebus")

# labels
plt.title("Data Penjualan Kacang")
plt.xlabel("Bulan")
plt.ylabel("Nilai Penjualan")
plt.xticks(sumbu_x, bulan)
# plt.grid(True)
plt.legend()
plt.show()
```

## DataFrame untuk Bar Chart

```python
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
penjualan_a = [1000, 3500, 7000, 6000, 6500, 3600, 4500, 5000, 1500, 2000, 1300, 8000]
penjualan_b = [200, 500, 800, 800, 700, 900, 500, 450, 300, 130, 500, 1700]

df_sales = pd.DataFrame({"bulan": bulan, "kacang_goreng": penjualan_a, "kacang_rebus": penjualan_b})
df_sales
```

```python
sumbu_x = np.arange(len(bulan))

chart = plt.subplot()

chart.bar(x=sumbu_x - 0.2, height=df_sales["kacang_goreng"], width=0.4, label="Kacang Goreng")
chart.bar(x=sumbu_x - 0.4, height=df_sales["kacang_rebus"], width=0.4, label="Kacang Rebus")

# labels
plt.title("Data Penjualan Kacang")
plt.xlabel("Bulan")
plt.ylabel("Nilai Penjualan")
plt.xticks(sumbu_x, bulan)
# plt.grid(True)
plt.legend()
plt.show()
```
