# Line Plotting

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Membuat plot melalui list

```python
l_data = [45, 67, 30, 22, 90, 100, 120]
fig = plt.figure(figsize=(10, 5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan dalam Jutaan")
plt.plot(l_data)
plt.show()
```

## Subplot

```python
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
data1 = [45, 67, 30, 22, 90, 100]
data2 = [50, 69, 55, 30, 75, 90]

fig = plt.figure(figsize=(10, 5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan dalam Jutaan")
chart = plt.subplot()
chart.plot(bulan, data1)
chart.plot(data2)
plt.show()
```

## Line Style
- "-" or solid
- "--" or dashed
- "-." or dashdot
- ":" or dotted

```python
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
data1 = [45, 67, 30, 22, 90, 100]
data2 = [50, 69, 55, 30, 75, 90]

fig = plt.figure(figsize=(10, 5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan dalam Jutaan")
chart = plt.subplot()
chart.plot(bulan, data1, ls="dotted", marker=".", color="m")
chart.plot(data2, ls="--", marker="o", color="c")
plt.show()
```

## Multiple Subplot

```python
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
data1 = [45, 67, 30, 22, 90, 100]
data2 = [50, 69, 55, 30, 75, 90]

fig = plt.figure(figsize=(10, 10))

plt.xlabel("Bulan")
plt.ylabel("Penjualan dalam Jutaan")

chart1 = plt.subplot(311)
chart2 = plt.subplot(312)

chart1.set(title="Data Penjualan A")
chart2.set(title="Data Penjualan B")

chart1.plot(bulan, data1, ls="dotted", marker=".", color="r")
chart2.plot(data2, ls="dashed", marker="o", color="b")

plt.show()
```

# Line Chart dengan Plotly

```python
import plotly.express as px
```

```python
nomor = [1, 2, 3, 4, 5]
data_penjualan = [300, 200, 150, 500, 700]

chart = px.line(x=nomor, y=data_penjualan, title="Data Penjualan", width=800, height=800)
chart.show()
```

```python
# Membuat plot dengan dictionary
data = {"nomor": [1, 2, 3, 4, 5], "data_penjualan": [300, 200, 150, 500, 700]}
df = pd.DataFrame(data)

chart = px.line(df, x="nomor", y="data_penjualan", title="Data Penjualan")
chart.show()
```
