### Sort Values

```python
df_sort_mvalues = df.sort_values(by="market_value")
df_sort_mvalues
```

```python
df_sort_mvalues = df.sort_values(by="market_value", ascending=False)
df_sort_mvalues
```

```python
pbola_usia_30 = df["age"] < 30
pbola_mvalue_50 = df["market_value"] > 50
df_pbola = df[(pbola_usia_30) & (pbola_mvalue_50)].sort_values(by="market_value", ascending=False)
df_pbola
```

## Mengcopy data set dan mengambil sebagian data

```python
df_small = df.iloc[:5,0:3]
df_small
```

## Menambah kolom

```python
nickname = pd.Series(["Alex", "Mesut", "Petr", "Theo", "Laurent"])
df_small.insert(0, "nickname", nickname)
df_small
```

```python
# menambahkan kolom dengan metode assign
df_small = df_small.assign(goals=[15, 20, 18, 11, 13])
df_small
```

```python
# menghapus kolom
# menghapus kolom nickname
df_small = df_small.drop(columns=["nickname"])
df_small
```

```python
# menambahkan data baru/row baru dengan pd.concat
endang = pd.Series({"name": "Endang Ismaya", "club": "Real Madrid", "age": 25, "goals": 65})
df_small = pd.concat([df_small, pd.DataFrame([endang])], ignore_index=True)
```

```python
alde = pd.Series({"name": "Aldeind Yussuf", "club": "Real Madrid", "age": 20, "goals": 25})
auza = pd.Series({"name": "Auza Tsabit", "club": "Chelsea", "age": 18, "goals": 5})
df_small = pd.concat([df_small, pd.DataFrame([alde, auza])], ignore_index=True)
df_small
```
